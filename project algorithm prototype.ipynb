```python
import networkx as nx
import matplotlib.pyplot as plt
G=nx.Graph()
```

```python
#add nodes from 0 to 7
G.add_nodes_from([i for i in range(8)])
```

```python
nx.draw(G,with_labels=1)
plt.plot()
```

```python
#add edges
G.add_edge(0,1)
G.add_edge(0,2)
G.add_edge(2,3)
G.add_edge(0,3)
G.add_edge(0,7)
G.add_edge(3,4)
G.add_edge(3,5)
G.add_edge(5,7)
G.add_edge(5,6)
G.add_edge(6,7)
G.add_edge(4,5)
G.add_edge(7,1)
#number of edges
nx.number_of_edges(G)
```

```python
nx.draw(G,with_labels=1)
plt.plot()
```

```python
#average shortest path
nx.average_shortest_path_length(G)
```

```python
#average clustering
nx.average_clustering(G)
```

```python
#get the list of neighbors of 0 in list a
a=list(G.neighbors(0))
```

```python
#Then remove node 0. Remember, the edges of nodes which are connected to node 0 are also lost.
G.remove_node(0)
```

```python
#Add the edges between the neighbors of 0. Same number of edges are added again. 
for i in a:
    for j in a:
        #if node i is different from node j and node i,j don't have an edge.
        if i!=j and G.has_edge(i,j)==False:
            #then add the edge here
            G.add_edge(i,j)
            
```

```python
nx.draw(G,with_labels=1)
plt.show()
```

```python
#metrics after the node 0 is removed and edges are added.
nx.average_clustering(G)
```

```python
nx.number_of_edges(G)
```

```python
nx.average_shortest_path_length(G)
```
